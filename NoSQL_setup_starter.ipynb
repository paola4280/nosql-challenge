5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
from bson import Decimal128
from decimal import Decimal

query = {'geocode.latitude': {'$type': 'string'}, 'geocode.longitude': {'$type': 'string'}}
documents = establishments.find(query)

for document in documents:
    latitude_str = document['geocode']['latitude']
    longitude_str = document['geocode']['longitude']
  
    latitude_decimal = Decimal128(Decimal(latitude_str))
    longitude_decimal = Decimal128(Decimal(longitude_str))
    
    update = {
        '$set': {
            'geocode.latitude': latitude_decimal,
            'geocode.longitude': longitude_decimal
        }
    }
    
    establishments.update_one({'_id': document['_id']}, update)

print("Latitude and longitude updated to Decimal128.")
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue
query = {'RatingValue': {'$type': 'string'}}
documents = establishments.find(query)

for document in documents:
    rating_value_str = document['RatingValue']

    try:
        rating_value_int = int(rating_value_str)
    except ValueError:
        # Handle any conversion errors if the string is not a valid number
        print(f"Skipping document with invalid RatingValue: {rating_value_str}")
        continue
    
    update = {
        '$set': {'RatingValue': rating_value_int}
    }
    
    establishments.update_one({'_id': document['_id']}, update)

print("RatingValue updated to Integer.")
```

```python
# Check that the coordinates and rating value are now numbers
document = establishments.find_one()
latitude_type = type(document['geocode']['latitude'])
longitude_type = type(document['geocode']['longitude'])
rating_value_type = type(document['RatingValue'])

print(f"Latitude type: {latitude_type}")
print(f"Longitude type: {longitude_type}")
print(f"RatingValue type: {rating_value_type}")
```
